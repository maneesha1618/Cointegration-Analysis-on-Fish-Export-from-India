```python
# to check the stationarity
from statsmodels.tsa.stattools import adfuller
price_column = df1['Price']
result=adfuller(price_column)
adf_statistics = result[0]
p_value = result[1]
```

```python
print(f'ADF Statistics: {adf_statistics}')
print(f'p_value: {p_value}')
```

```python
# result
significance_level =0.05
if p_value < significance_level:
    print("The time series is stationary (reject the null hypothesis).")
else:
    print("The time series is non-stationary (fail to reject the null hypothesis).")
```

```python
df1['adj_log'] = np.log(df1['Price'])
df1['adj_sqrt'] = np.sqrt(df1['Price'])
df1['adj_cbrt'] = np.cbrt(df1['Price'])
```

```python
df1 = df1.dropna()
df1
```

```python
df1['Price_log_diff'] = df1['adj_log'] - df1['adj_log'].shift(1)
```

```python
df1 = df1.dropna()
df1
```

```python
# again checking the stationarity  after  log transformation

from statsmodels.tsa.stattools import adfuller
price_column = df1['Price_log_diff']
result=adfuller(price_column)
adf_statistics = result[0]
p_value = result[1]
print(f'ADF Statistics: {adf_statistics}')
print(f'p_value: {p_value}')
# result
significance_level = 0.05
if p_value < significance_level :
    print("The time series is  stationary (reject the null hypothesis).")
else:
    print("The time series is non-stationary (fail to reject the null hypothesis).")
```
